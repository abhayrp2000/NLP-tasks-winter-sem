```python
all_text
```

```python
import nltk
```

```python
tokenized_sent=nltk.word_tokenize(all_text)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
words= [w for w in tokenized_sent if not w in stop_words]
```

```python
tagged_words=nltk.pos_tag(words)
```

```python
tagged_words
```

```python
adjectives=[]

for i in range(0,len(tagged_words)):
    if(tagged_words[i][1]=='JJ'):
        adjectives.append(tagged_words[i][0])
        
```

```python
adjectives
```

```python
import collections
freq_words=collections.Counter(adjectives)
```

```python
freq_words
```

```python
res=freq_words.most_common(15)
res
```
